### -------------------------------------------------------------------------------------

### "Location Full" va "Institution Name" ustunlari birga indeks bolib yaratildi DataFrame da har bir qator unik yagona (Location Full, Institution Name) juftligiga ega bo'ladi

```python
name_loc_ind = df.set_index(["Location Full", "Institution Name"])
name_loc_ind
```

### -------------------------------------------------------------------------------------

### Bu kodda loc indeks orqali ma'lum bir indeks qiymatlarini tanlab olish uchun ishlatiladi va "Uzbekistan" va "United States" qiymatiga ega bo'lgan barchasi chiqadi

```python
name_loc_ind.loc[["Uzbekistan", "United States"]]
```

### -------------------------------------------------------------------------------------


### Bu kodda loc indeks orqali ma'lum bir indeks qiymatlarni tanlab olish uchun ishlatiladi. Natijada ("Uzbekistan", "L") va ("Turkey", "M") kombinatsiyalariga ega bo'lgan barcha qatorlar chiqadi

```python
name_size_ind = df.set_index(["Location Full", "Size"])
name_size_ind.loc[[("Uzbekistan", "L"), ("Turkey", "M")]]
```

### -------------------------------------------------------------------------------------

### Location va  Universitetlarning nomi saqlab turgan ozgaruvchichi sortlavordik bu avval locationni indexni sortlaydi keyin esa Institution name index boyicha sortlaydi , masalan A birinchi bolgani uchun Argentina chiqdi va Argentinada mavjud Universitetlar Location barchasi sortlangandan keyin , Institution name lar sortlandi ya'ni Argentinadagi Universitetlar nomi sortlandi.

```python
name_loc_ind.sort_index()
```

### -------------------------------------------------------------------------------------


### Quyidagi kodda Birinchi index teskari sortlanadi ikkinchisi esa togri sortlanadi chunki biz ascending orqali belgiladik False bersak teskari tartiblaydi alfavit boyicha

```python
name_loc_ind.sort_index(level=["Location Full", "Institution Name"], ascending=[False, True])
```

### -------------------------------------------------------------------------------------

### Yuqoridagi kodga qo'shimcha ravishda qirqish amalidan foydalandik ya'ni [2:8] bu index 2dan boshlab 8chi indexgacha qirqib yangi dataframe chiqarib beradi.

```python
name_loc_ind.sort_index(level=["Location Full", "Institution Name"], ascending=[False, True])[2:8]
```

### -------------------------------------------------------------------------------------

### Bu kodda location va univer nomlarini index sifatida oldik u columlar dataframedan olib tashlanadi va indexlarni sortlavordik, avval Location keyin univer nomlari sortlanadi, keyin esa qirqish amalidan foydalanib Austriadan bahraingacha bolganlarini chiqardik 

```python
loc_srt = df.set_index(["Location Full", "Institution Name"]).sort_index()
loc_srt["Austria":"Bahrain"]
```

### -------------------------------------------------------------------------------------

### Quyidagi kodda Univer nomlari hamda Sizeni sortladik va United Statesni faqat Size L bolganlaridan boshlab O'zbekiston univerlarini size L bo'lganlarigacha chiqaradi

```python
name_size_srt = name_size_ind.sort_index()
name_size_srt.loc[("United States", "L"):("Uzbekistan", "L")]
```

### -------------------------------------------------------------------------------------

### bu kodda esa columnlarni Univer nomidan boshlab Faculty Student columngacha qirqib oldik va chiqardik

```python
name_size_srt.loc[:, "Institution Name":"Faculty Student"]
```

### -------------------------------------------------------------------------------------

### Bu kodda yuqoridagi ikkita kodni birlashitirilgani ya'ni Location United States va Size L bo'ganlaridan boshlab Unzbekistan size L bo'lganlarigacha bo'ladi , columnlar esa "Institution Name":"Faculty Student" -- Institution Name dan "Faculty Student" gacha chiqadi 

```python
name_size_srt.loc[("United States", "L"):("Uzbekistan", "L"), "Institution Name":"Faculty Student"]
```

### -------------------------------------------------------------------------------------

### "df" DataFrame dan pivot_table yordamida "Institution Name" ustunini indeks sifatida va "Size" ustunini ustunlar sifatida oladi "International Students" ustunini qiymat sifatida boladi keyin  intern_students_by_size_vs_name nomli yangi DataFrame yaratildi

### -------------------------------------------------------------------------------------
